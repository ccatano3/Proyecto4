```
from pyspark.ml.feature import Tokenizer, RegexTokenizer
from pyspark.sql.functions import col, udf
tokenization=Tokenizer(inputCol='content',outputCol='tokens')
regex_tokenizer= RegexTokenizer(inputCol='content', outputCol='tokens',pattern='[^a-zA-Z1-9]+').setMinTokenLength(2)
```

```
df=df.fillna("unknown", subset=["content","title"])
rg_tokenized = regex_tokenizer.transform(df)
```

```
rg_tokenized.show()
```

```
rg_tokenized.withColumn('tokens1', count_tokens(col('tokens'))).show()
```

```
from pyspark.ml.feature import StopWordsRemover
```

```
remover = StopWordsRemover(inputCol='tokens', outputCol='filtered')
```

```
refined_df=remover.transform(rg_tokenized)
```

```
refined_df.select(['title', 'tokens', 'filtered']).show(5)
```

```
from pyspark.ml.feature import CountVectorizer
count_vec=CountVectorizer(inputCol='filtered',
                          outputCol='features')
cv_df=count_vec.fit(refined_df).transform(refined_df)
cv_df.select(['title','filtered','features']).show(5)
```

```
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```
stop_words = set(stopwords.words('english'))
```

```
def clean(row):
  # Remove special characters
  _nrowp = str(row.content).lower() + str(row.title).lower()
  _nrow = ''.join(jiu for jiu in _nrowp if (jiu.isalpha() or jiu == ' '))
  # Get all group of words
  _nrow = _nrow.split()
  return [word for word in _nrow if (not word in stop_words and len(word) >= 2)]
```

```
def funny(res):
  key, value = res
  cur_sum = 0
  for (idnew, ntimes) in value:
    cur_sum += int(ntimes)
  return -cur_sum
```

```
def even(res):
  key, val = res
  cur_sum = 0.0
  for (idnew, ntimes) in val:
    cur_sum += int(ntimes)
  return [key, cur_sum]
```

```
rdd = useful_dataset.rdd.flatMap(lambda row: map(lambda word: (word, str(row._c0)), clean(row))).map(lambda text: ((text[0], text[1]), 1)).reduceByKey(lambda c1, c2: c1 + c2).map(lambda _key: (_key[0][0], (_key[0][1], _key[1]))).aggregateByKey(set(), fun, lambda x, y: x | y).sortBy(lambda res: funny(res))
res = rdd.collect()
```
